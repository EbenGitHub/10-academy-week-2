```python
# Task 2

# Aggregate metrics per customer
engagement_metrics = data.groupby('MSISDN/Number').agg({
    'Dur. (ms)': 'sum',  # Total session duration
    'HTTP DL (Bytes)': 'sum',  # Total download data
    'HTTP UL (Bytes)': 'sum',  # Total upload data
})

# Calculate total traffic (Download + Upload)
engagement_metrics['Total Traffic (Bytes)'] = engagement_metrics['HTTP DL (Bytes)'] + engagement_metrics['HTTP UL (Bytes)']

# Add session frequency
engagement_metrics['Session Frequency'] = data['MSISDN/Number'].value_counts()

# Rename columns for clarity
engagement_metrics.rename(columns={
    'Dur. (ms)': 'Total Duration (ms)',
    'HTTP DL (Bytes)': 'Total Download (Bytes)',
    'HTTP UL (Bytes)': 'Total Upload (Bytes)',
}, inplace=True)

engagement_metrics.head()
```

```python
# Top 10 customers by each metric
top_customers_duration = engagement_metrics.nlargest(10, 'Total Duration (ms)')
top_customers_download = engagement_metrics.nlargest(10, 'Total Download (Bytes)')
top_customers_upload = engagement_metrics.nlargest(10, 'Total Upload (Bytes)')
top_customers_traffic = engagement_metrics.nlargest(10, 'Total Traffic (Bytes)')

print("Top 10 Customers by Total Duration:")
print(top_customers_duration)
print("Top 10 Customers by Total Traffic:")
print(top_customers_traffic)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Select relevant columns for clustering
clustering_data = engagement_metrics[['Total Duration (ms)', 'Total Traffic (Bytes)', 'Session Frequency']]

# Normalize data
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(clustering_data)

print("Normalized Data:")
print(normalized_data[:5])
```

```python
from sklearn.cluster import KMeans

# Elbow method to find optimal k
inertia = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42).fit(normalized_data)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(range(1, 10), inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()

# K-Means Clustering (k=3)
kmeans = KMeans(n_clusters=3, random_state=42)
engagement_metrics['Cluster'] = kmeans.fit_predict(normalized_data)

print("Engagement Clusters:")
print(engagement_metrics['Cluster'].value_counts())
```

```python
# Visualize clusters using pairplot
engagement_metrics['Cluster'] = engagement_metrics['Cluster'].astype(str)  # Convert to categorical
sns.pairplot(engagement_metrics, hue='Cluster', diag_kind='kde', palette='Set2')
plt.show()
```

```python
cluster_stats = engagement_metrics.groupby('Cluster').mean()
print("Cluster Statistics:")
print(cluster_stats)
```

```python
# Task 3
# Aggregate metrics per customer
experience_metrics = data.groupby('MSISDN/Number').agg({
    'TCP DL Retrans. Vol (Bytes)': 'mean',  # Average TCP retransmission
    'Avg RTT DL (ms)': 'mean',  # Average Round Trip Time (RTT)
    'Avg Bearer TP DL (kbps)': 'mean',  # Average throughput
    'Handset Type': 'first',  # Handset type
}).reset_index()

# Rename columns for clarity
experience_metrics.rename(columns={
    'TCP DL Retrans. Vol (Bytes)': 'Avg TCP Retrans (Bytes)',
    'Avg RTT DL (ms)': 'Avg RTT (ms)',
    'Avg Bearer TP DL (kbps)': 'Avg Throughput (kbps)'
}, inplace=True)

experience_metrics.head()
```

```python
# Top 10 values
top_tcp_retrans = experience_metrics.nlargest(10, 'Avg TCP Retrans (Bytes)')
top_rtt = experience_metrics.nlargest(10, 'Avg RTT (ms)')
top_throughput = experience_metrics.nlargest(10, 'Avg Throughput (kbps)')

# Bottom 10 values
bottom_tcp_retrans = experience_metrics.nsmallest(10, 'Avg TCP Retrans (Bytes)')
bottom_rtt = experience_metrics.nsmallest(10, 'Avg RTT (ms)')
bottom_throughput = experience_metrics.nsmallest(10, 'Avg Throughput (kbps)')

# Most frequent values
most_freq_tcp_retrans = experience_metrics['Avg TCP Retrans (Bytes)'].mode()
most_freq_rtt = experience_metrics['Avg RTT (ms)'].mode()
most_freq_throughput = experience_metrics['Avg Throughput (kbps)'].mode()
```

```python
# Distribution of Average Throughput per Handset Type
sns.boxplot(x='Handset Type', y='Avg Throughput (kbps)', data=experience_metrics)
plt.title('Distribution of Average Throughput per Handset Type')
plt.xticks(rotation=45)
plt.show()

# Distribution of Average TCP Retransmission per Handset Type
sns.boxplot(x='Handset Type', y='Avg TCP Retrans (Bytes)', data=experience_metrics)
plt.title('Distribution of Average TCP Retransmission per Handset Type')
plt.xticks(rotation=45)
plt.show()
```

```python
# Normalize the data
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(experience_metrics[['Avg TCP Retrans (Bytes)', 'Avg RTT (ms)', 'Avg Throughput (kbps)']])

# K-Means Clustering (k=3)
kmeans = KMeans(n_clusters=3, random_state=42)
experience_metrics['Cluster'] = kmeans.fit_predict(normalized_data)

# Visualize Clusters
sns.pairplot(experience_metrics, hue='Cluster', diag_kind='kde', palette='Set2')
plt.title('User Experience Clusters')
plt.show()
```
